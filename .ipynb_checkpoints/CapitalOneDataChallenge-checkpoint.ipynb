# Problem Statement: 
# To analyze the given data to identify the zipcodes in New York that are worth investing for most profit on short term rentals

### PROBLEM DESCRIPTION
The real estate company wants to specifically invest in New York City. The investors have already concluded that two bedroom properties are most profitable. We need to analyze the given data and suggest them zipcodes within the city that would be most profitable with less amount of risk.  

### ASSUMPTIONS 
1. We assume that all the data from data sources (Airbnb and Zillow) is correct in terms of values, zipcodes and their corresponsing cities

2. Prices of properties are not affected by seasonality.

3. All the properties are in good condition and there is no significant repair or maintaince charges

4. Tax is constant across all the zipcodes

5. The investor will pay for the property in cash (i.e. no mortgage/interest rate will need to be accounted for). 

6. The time value of money discount rate is 0% (i.e. $1 today is worth the same 100 years from now).

7. All properties and all square feet within each locale can be assumed to be homogeneous (i.e. a 1000 square foot property in a locale such as Bronx or Manhattan generates twice the revenue and costs twice as much as any other 500 square foot property within that same locale.)

One more observation is that we are given the latitude and longitude for the given areas within New York City. Using these two parameters we can find the missing zipcodes . 

The function below will help us get rid of the missing values.

```python
# we write a function to extract the latitude and longitude 
def convert_zipcode(lat,long):
    search = SearchEngine(simple_zipcode=True)
    result = search.by_coordinates(lat,long, radius=10, returns=1)    
    res=result[0]
    return res.zipcode

#convert zipcode to type string and limit it to 5 
airbnb['zipcode'] = airbnb['zipcode'].astype('str').apply(lambda x:x[:5])
cleanzipcode =airbnb[~airbnb['zipcode'].isin(['nan'])] 
missingzip = airbnb[airbnb['zipcode'].isin(['nan'])]

#print length of zipcode
print('The lenght of the zipcodes is ', len(airbnb.zipcode))

#length of zipocodes that are present 
print('The lenght of the zipcodes present excluding null values is ', len(cleanzipcode.zipcode))
# length of missing values in zipcode
print('The lenght of the zipcodes with null values is ', len(missingzip.zipcode))

#finding zipcodes based on the langitude and latitude given in the dataset 
missingzip.zipcode = missingzip.apply(lambda row: convert_zipcode(row['latitude'], row['longitude']), axis=1)

#again print the length of missing values in this case will be zero
print(len(missingzip[missingzip.zipcode=='nan']))

print(cleanzipcode.zipcode.unique())# print all unique zipcodes

print(missingzip.zipcode.unique())# newly acquired zipcodes

# append the previous and newly acquired zipcodes
airbnb_clean = cleanzipcode.append(missingzip, ignore_index=True)

print('The length on newly acquired zipcodes matches to the original length of zipcodes ie',len(airbnb.zipcode))
```

# 4. Price per Night 

We now analyze the price per night for each zipcode within New York City.
Beyond investment, company would look foward to earn profit. Higher the rental price, faster is the ROI and profits.

```python
fig_out = plt.figure(figsize=(15,8))
ax1 = fig_out.add_subplot(111)
result_df.boxplot(column='price',by='zipcode',ax=ax1)
ax1.tick_params('x',labelrotation = 90)
ax1.set_ylabel("Airbnb Daily Price",fontsize = 20)
ax1.set_xlabel("Zipcodes", fontsize = 20)
```

Price per day are ridiculously high for some zipcodes.  There are outliers as there are not included in the box of other observation i.e no where near the quartiles.

We now use the IQR method to get rid of the outliers. Interquartile range (IQR), also called the midspread or middle 50%, or technically H-spread, is a measure of statistical dispersion, being equal to the difference between 75th and 25th percentiles, or between upper and lower quartiles, IQR = Q3 − Q1.
                            
                            𝐼𝑄𝑅 = 𝑄3 ( 75𝑡h 𝑝𝑒𝑟𝑐𝑒𝑛𝑡𝑖𝑙𝑒) − 𝑄1 (25𝑡h 𝑝𝑒𝑟𝑐𝑒𝑛𝑡𝑖𝑙𝑒)

Reference: https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba


```python
# use IQR score to remove outliers
Q1 = result_df.price.quantile(0.25)
Q3 = result_df.price.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
result_df_clean = result_df[~((result_df.price < (Q1 - 1.5 * IQR)) |(result_df.price > (Q3 + 1.5 * IQR)))]
result_df_clean.shape
```

```python
fig_out = plt.figure(figsize=(15,8))
ax1 = fig_out.add_subplot(111)
result_df_clean.boxplot(column='price',by='zipcode',ax=ax1)
ax1.tick_params('x',labelrotation = 90)
ax1.set_ylabel("Airbnb Daily Price",fontsize = 20)
ax1.set_xlabel("Zipcodes", fontsize = 20)
```

###  The listing with price that is not in this range would be dropped.

# 5. Occupancy Rate Analysis

Although it is given that we can assume the occupancy rate to be 75%, generalizing it would be not that accurate and it is obvious that not all property will have the same occupancy rate.

The review score of a property has a great impact on occupancy rate. Since there is no real occupancy rate for each property, we took the data from Mashvisor, which is a well-known real estate platform offering metadata in real estate market and build an occupancy rate table based on the review_scores_location from revenue data.

Reference: https://www.mashvisor.com/blog/breaking-even-real-estate-investment-property/


Lower the availablity , higher the occupancy - indicating fast-filling properties.

```python
occ_table = pd.DataFrame({'Location_Score':['9.5-10','8.5-9.5','7.5-8.5','6.5-7.5','<6.5'],
                          'Occupancy_Rate':[0.8, 0.75, 0.7, 0.65, 0.5]})
data=result_df_clean['review_scores_location']
condlist = [(data<=10) & (data>9.5), (data<=9.5) & (data>8.5), (data<=8.5) & (data>7.5), \
                (data<=7.5) & (data>6.5), (data<= 6.5)]
choicelist = [0.8, 0.75, 0.7, 0.65, 0.5]
result_df_clean['occupancy_rate'] = np.select(condlist,choicelist, default = 0)
```

```python
fig, ax= plt.subplots(figsize=(20,10))
occupancyfig = sns.barplot(x='zipcode',y='occupancy_rate',hue='neighbourhood_group_cleansed', 
                           data = result_df_clean, ci= None)
sns.set(style="dark")
plt.xlabel("Zipcode", fontsize = 20)
plt.ylabel("Occupancy Rate", fontsize = 20)
plt.title("Occupancy Rate Analysis", fontsize = 20)
plt.show(occupancyfig)
```

# 7 - Expected return 

Expected return is the net income during a period. We calculated the expected return based on such formula.

### 𝐸𝑥𝑝𝑒𝑐𝑡𝑒𝑑 𝑅𝑒𝑡𝑢𝑟𝑛 = 𝐷𝑎𝑖𝑙𝑦 𝐿𝑖𝑠𝑡𝑖𝑛𝑔 𝑃𝑟𝑖𝑐𝑒 × 𝑇𝑖𝑚𝑒 (365 𝑑𝑎𝑦𝑠) × 𝑂𝑐𝑐𝑢𝑝𝑎𝑛𝑦 𝑅𝑎𝑡𝑒 

```python
result_df_clean['expected_return']= ((result_df_clean['price'])*365*(result_df_clean['occupancy_rate']))
```

```python
fig, ax= plt.subplots(figsize=(20,10))
roifig = sns.barplot(x='zipcode',y='expected_return',hue='neighbourhood_group_cleansed', data =result_df_clean, palette='husl', ci =None)
sns.set(style="dark")
plt.xlabel("Zipcode", fontsize = 20)
plt.ylabel("Expected Return", fontsize = 20)
plt.title("EXPECTED RETURN ON INVESTMENT", fontsize = 20)
plt.show(roifig)
```

# 8. Breakeven Analysis

To better understand how long it will take to break even when we buy a property within New York City, we need to decide the break-even point. The definition of the break-even point is the point where total revenues equal to the costs. Since one of the assumptions we have are 0% discount rate and pay the property in cash, we can simply our formula as follow: 
# 1/ Return Ratio

It is the reciprocal of Return Ratio

```python
result_df_clean['return_ratio']= (((result_df_clean['price']*365*(result_df_clean['occupancy_rate'])))/result_df_clean['estimated_price'])
result_df_clean['breakeven']= 1/ result_df_clean['return_ratio']
```

```python
temp = result_df_clean[['breakeven','zipcode']]
result_df_clean= result_df_clean.replace([np.inf, -np.inf], np.nan).dropna(subset=["breakeven"], how="all")
```

```python
fig, ax= plt.subplots(figsize=(20,10))
costfig = sns.barplot(x='zipcode',y='breakeven',hue='neighbourhood_group_cleansed', data =result_df_clean, palette='husl', ci = None)
sns.set(style="dark")
plt.xlabel("Zipcode", fontsize=20)
plt.ylabel("BREAKEVEN", fontsize=20)
plt.title("BREAKEVEN ANALYSIS", fontsize=20)
plt.show(costfig)
```

Giving Rank to Zipcode for all Analysis and Choosing the best location
With the above all preliminary analysis, we have got various different opinions on which could be better on different terms.

So, we need a measure to cumulate all the results to look at the most profitable investment.

Here, I have ranked the zipcodes for every other analysis. The best value is given the minimal rank (Rank 1), and the worst the highest.

So, when we aggregate using mean, the most minimal rank will be our winner.

```python
# Groupby zipcode
rank=result_df_clean.groupby('zipcode')

# Considering the parameters of Interest 
rank1=rank[['price', 'estimated_price', 'number_of_reviews_x', 'review_scores_location', 'return_ratio', 'occupancy_rate']].mean()

# Setting IndexO
rank1=rank1.reset_index()

# Giving Ranks
rank1['Price per Day'] = rank1['price'].rank(ascending=False)
rank1['Buying Cost'] = rank1['estimated_price'].rank(ascending=True)
rank1['Number of Reviews'] = rank1['number_of_reviews_x'].rank(ascending=False)
rank1['Location Review Score'] = rank1['review_scores_location'].rank(ascending=False)
rank1['Occupancy Rate Score'] = rank1['occupancy_rate'].rank(ascending=False)
rank1['Return Ratio Rank'] = rank1['return_ratio'].rank(ascending=False)


# Total Rank
rank1 ['Total Rank Score'] = (rank1['Price per Day']+ rank1['Buying Cost'] + rank1['Number of Reviews']
                              + rank1['Location Review Score'] + rank1['Occupancy Rate Score'] + rank1['Return Ratio Rank'])/6

# Merge Rank1 to Main Table 
result_df_clean = rank1.merge(result_df_clean, how='left', on='zipcode')

# Visualize Rank1
rank1
```

```python
fig, ax= plt.subplots(figsize=(20,10))
rankfig = sns.barplot(x='zipcode',y='Total Rank Score',hue='neighbourhood_group_cleansed', data =result_df_clean, palette='husl')
sns.set(style="whitegrid")
plt.xlabel("Zipcode")
plt.ylabel("Cumulative Rank")
plt.title("Cumulative Analysis based on Rank")
plt.show(rankfig)
```

```python
# Groupby zipcode
final=result_df_clean.groupby('zipcode')

# Considering the parameters of Interest 
best =final[['Total Rank Score']].mean()

# Setting Index
best=best.reset_index()

# Top 5 Zipcodes to Invest based on the rank
best.sort_values('Total Rank Score', ascending = True)
```
